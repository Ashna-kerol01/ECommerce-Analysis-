7. Factors Behind High Return Rates

Correlation Between Return Rate and Pricing Factors 

```python
return_corr = df[
    ["Price", "Discount", "Shipping Cost", "Tax Rate", "Return Rate"]
].corr()

sbn.heatmap(return_corr, annot=True, cmap="coolwarm")
plt.title("Correlation with Return Rate")
plt.show()
```

## ðŸ“Š Correlation Analysis â€“ Summary

- ðŸ” *Return rate has no strong correlation* with price, discount, shipping cost, or tax rate.  
- âž– All relationships are *very weak or near zero*.  
- ðŸ§  Returns are likely driven by *non-price factors* such as product quality and delivery experience.  

### ðŸŽ¯ Conclusion
- Pricing-related variables *do not significantly influence returns*.  
- Further EDA should focus on *categorical features* (shipping method, category, seasonality).

9. Supplier Performance Comparison

```python
supplier_perf = df.groupby("Supplier ID").agg({
    "Price": "mean",
    "Stock Level": "mean",
    "Return Rate": "mean",
    "Popularity Index": "mean"
}).sort_values("Popularity Index", ascending=False)

supplier_perf
supplier_perf.head()

```

```python
sbn.scatterplot(
    data=supplier_perf,
    x="Price",
    y="Popularity Index",
    size="Return Rate",
    sizes=(50, 300)
)
plt.title("Supplier Performance Overview")
plt.show()
```

## ðŸ“Œ Supplier Performance â€” Summary

- ðŸ’° *Moderate pricing (60â€“80)* is associated with *higher popularity* among suppliers.
- ðŸš« *High price does not guarantee high popularity*, indicating price sensitivity.
- ðŸ”„ *Higher return rates* are mostly seen in *mid-to-high priced suppliers*, even if they are popular.
- â­ *Best-performing suppliers* fall within a balanced zone of *moderate price, high popularity, and low return rates*.
- âš ï¸ *Outliers with high price and high return rate* represent potential risk and require attention.
- ðŸ§  *Business Insight:* Optimizing supplier selection based on balanced pricing and return rate control can improve overall performance.

 10. Key Drivers of Product Popularity (Correlation & Multivariate)


```python
popularity_corr = df[
    ["Price", "Discount", "Tax Rate", "Stock Level", "Shipping Cost", "Return Rate", "Popularity Index"]
].corr()

sbn.heatmap(popularity_corr, annot=True, cmap="coolwarm")

plt.title(" Popularity Drivers")
plt.show()
#Correlation Matrix â€“
```

## ðŸ“Œ Correlation Analysis â€” Summary

- ðŸ“‰ All variables show *very weak correlations*, indicating no strong linear relationships.
- ðŸ’° Price and ðŸ·ï¸ Discount have *negligible impact* on Popularity Index.
- ðŸ”„ Return Rate is largely *independent* of other factors.
- ðŸ­ Operational variables (Stock Level, Shipping Cost, Tax Rate) show *minimal influence* on popularity.
- ðŸ§  *Key Insight:* Popularity is likely driven by qualitative or non-linear factors rather than individual numerical variables.
